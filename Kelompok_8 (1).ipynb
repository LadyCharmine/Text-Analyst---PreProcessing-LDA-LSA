# Cleaning

```python
import string
import re #regex library
```

```python
# ------ Cleaning ---------

nltk.download('punkt')

def remove_tweet_special(text):
    # remove tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # remove non ASCII (emoticon, chinese word, .etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # remove mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove incomplete URL
    return text.replace("http://", " ").replace("https://", " ")

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_tweet_special)

#remove number
def remove_number(text):
    return  re.sub(r"\d+", "", text)

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_number)

#remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_punctuation)

#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_whitespace_LT)

#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_whitespace_multiple)

# remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

uinsa_clean['abstrak'] = uinsa_clean['abstrak'].apply(remove_singl_char)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
lsa = TruncatedSVD(n_components=10, n_iter=10, random_state=42)
```

```python
lsa_matrix = lsa.fit_transform(bow_df)
```

```python
print(bow_df.shape) #hidden
print(lsa_matrix.shape) #wight / code
print(lsa.components_.shape) #fitur/topic
```

```python
def get_topic(model, vocab) :
    return [[vocab[idx] for idx in comp.argsort()[-6:][::-1] if vocab[idx].isalnum()]
            for comp in model.components_]
```

```python
topics = get_topic(lsa, vocab)
topics
```

```python
from sklearn.metrics import pairwise_distances
from sklearn.metrics import silhouette_score
from sklearn.metrics import classification_report
```

```python
def topic_coherence(model, vectors):
    topic_coherence_scores = []
    for i in range(model.components_.shape[0]):
        topic_terms = [vocab[idx] for idx in model.components_[i].argsort()[-6:][::-1] if vocab[idx].isalnum()]
        topic_vectors = vectors[:, i]  # Mengubah baris ini
        pairwise_dists = pairwise_distances(topic_vectors[:, np.newaxis], metric='cosine')  # Mengubah baris ini
        avg_coherence = 0
        for j in range(len(topic_terms)):
            coherence_score = (pairwise_dists.sum() - pairwise_dists[j, :].sum()) / (len(topic_terms) - 1)
            avg_coherence += coherence_score
        avg_coherence /= len(topic_terms)
        topic_coherence_scores.append(avg_coherence)
    return topic_coherence_scores
```

```python
topic_coherence_scores = topic_coherence(lsa, lsa_matrix)
print("Topic Coherence Scores:")
for i, score in enumerate(topic_coherence_scores):
    print("Topic", i+1, ":", score)
```

```python
#  Mengukur Topic Significance Metrics
def topic_significance(model):
    topic_significance_scores = []
    total_variance = np.sum(model.explained_variance_)
    for i in range(model.components_.shape[0]):
        topic_variance = model.explained_variance_[i]
        topic_significance = topic_variance / total_variance
        topic_significance_scores.append(topic_significance)
    return topic_significance_scores

topic_significance_scores = topic_significance(lsa)
print("Topic Significance Scores:")
for i, score in enumerate(topic_significance_scores):
    print("Topic", i+1, ":", score)
```

```python
# Mengukur Topic Diversity Metrics
def topic_diversity(model):
    topic_diversity_scores = []
    for i in range(model.components_.shape[0]):
        topic_vector = model.components_[i]
        topic_diversity = np.sqrt(np.sum(np.square(topic_vector)))
        topic_diversity_scores.append(topic_diversity)
    return topic_diversity_scores

topic_diversity_scores = topic_diversity(lsa)
print("Topic Diversity Scores:")
for i, score in enumerate(topic_diversity_scores):
    print("Topic", i+1, ":", score)
```

```python
import gensim
from gensim import corpora

dictionary = corpora.Dictionary(doc_clean)
print(dictionary)

doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
# Membuat objek untuk model LDA menggunakan pustaka gensim.
Lda = gensim.models.ldamodel.LdaModel

total_topics = 3 # jumlah topik yang akan di extract
number_words = 10 # jumlah kata per topik
```

```python
# Menjalankan dan Melatih model LDA pada matriks dokumen-istilah.
lda_model = Lda(doc_term_matrix, num_topics=total_topics, id2word = dictionary, passes=50)

lda_model.show_topics(num_topics=total_topics, num_words=number_words)
```

```python
# Jumlah Kata Kunci Topik

from collections import Counter
topics = lda_model.show_topics(formatted=False)
data_flat = [w for w_list in doc_clean for w in w_list]
counter = Counter(data_flat)

out = []
for i, topic in topics:
    for word, weight in topic:
        out.append([word, i , weight, counter[word]])

df_imp_wcount = pd.DataFrame(out, columns=['word', 'topic_id', 'importance', 'word_count'])
print(df_imp_wcount)
```

```python
#Topik dominan dan kontribusi persentase dalam setiap topik.
def format_topics_sentences(ldamodel=None, corpus=doc_term_matrix, texts=document):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list
        # print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)
```
